```python
import matplotlib
matplotlib.use('Agg')  # 使用非交互式后端，避免 matplotlib_inline 的兼容性问题
import matplotlib.pyplot as plt
```

```python
import numpy as np


index=4
step100 = np.load(f'/inspire/hdd/project/roboticsystem2/jingzhanghui-253108140204/world_model/RL/Epona/test_videos/test-traj-100steps/00{index}.npy')


step10 = np.load(f'/inspire/hdd/project/roboticsystem2/jingzhanghui-253108140204/world_model/RL/Epona/test_videos/test-traj-10steps/00{index}.npy')

step5_8samples = np.load(f'/inspire/hdd/project/roboticsystem2/jingzhanghui-253108140204/world_model/RL/Epona/test_videos/test-traj-5steps_8samples/000_traj01.npy')
```
